```python
plt.hist(x="Age",data=titanic,bins=7,edgecolor='blue')
plt.title('survival per age')
plt.ylabel('Survival counts')
plt.xlabel('Age')
```

```python
new=titanic.groupby('Pclass')['Survived'].mean()
new
```

```python
plt.pie(new,labels=[f'Class {i}' for i in new.index])
```

```python
survival=titanic.groupby('Survived').size()
```

```python
survival.plot.pie(y='Survived',title="Rate of survival")
```

```python
sex=titanic.groupby('Sex')['Survived'].mean()*100
sex
```

```python
plt.bar(x=sex,height='Survived')
```
